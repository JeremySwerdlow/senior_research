## Example Trees for Paper

```python
from tree import create_decision_tree, add_new_data
from graph_tree import graph_tree

import pandas as pd
```

```python
pet_df = pd.read_csv('../datasets/pets.txt', '\t')
idx = list(range(len(pet_df)))
idx.remove(4)
idx.remove(14)
pet_tree_first = create_decision_tree(pet_df.iloc[idx],
                                        list(pet_df.columns[:-1]),
                                        'iscat',
                                        pet_df)
pet_graph_first = graph_tree(pet_tree_first)
pet_graph_first
```

```python
pet_tree_new_node = add_new_data(pet_df[pet_df['size'] == 'enormous'], pet_tree_first)
pet_graph_new_node = graph_tree(pet_tree_new_node)
pet_graph_new_node
```

```python
pet_tree_full = add_new_data(pd.DataFrame(pet_df.iloc[4]).transpose(), pet_tree_new_node)
pet_graph_full = graph_tree(pet_tree_full)
pet_graph_full
```
